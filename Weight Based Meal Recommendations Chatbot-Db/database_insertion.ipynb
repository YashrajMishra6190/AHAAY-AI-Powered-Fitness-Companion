```python
import pymongo

def connect_mongodb():
    client = pymongo.MongoClient("mongodb://localhost:27017/")
    db = client["chatbot-db"]
    return db

def read_txt_file(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        lines = file.readlines()
    
    qa_pairs = []
    question, answer = None, None
    
    for line in lines:
        line = line.strip()
        if line.lower().startswith("question"):
            question = line.split(":", 1)[-1].strip()
        elif line.lower().startswith("answer"):
            answer = line.split(":", 1)[-1].strip()
            if question and answer:
                qa_pairs.append({"question": question, "answer": answer})
                question, answer = None, None
    
    return qa_pairs

def insert_data(db, collection_name, file_path):
    data = read_txt_file(file_path)
    if data:
        db[collection_name].insert_many(data)
        print(f"Inserted {len(data)} records into {collection_name}")
    else:
        print(f"No data found in {file_path}")

if __name__ == "__main__":
    db = connect_mongodb()
    
    files = {
        "Male_Underweight": "male_underweight.txt",
        "Male_Normalweight": "male_normalweight.txt",
        "Male_Overweight": "male_overweight.txt",
        "Male_Obesity": "male_obesity.txt",
        "Female_Underweight": "female_underweight.txt",
        "Female_Normalweight": "female_normalweight.txt",
        "Female_Overweight": "female_overweight.txt",
        "Female_Obesity": "female_obesity.txt"
    }
    
    for collection, filename in files.items():
        insert_data(db, collection, filename)

    print("Data insertion complete.")
```
